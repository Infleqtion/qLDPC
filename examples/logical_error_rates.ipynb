# Logical vs. physical error rates in a code capacity model

- Physical errors are sampled by flipping each bit (or qubit) with some probability.
- The logical error is the fraction of physical errors that are decoded incorrectly, resulting in a logical error.

```python
import matplotlib.pyplot as plt
import qldpc
from scripts.utils import make_error_rate_figure
```

## Plot logical vs. physical error rates for the classical repetition code and quantum surface code

Decode with minimum-weight perfect matching (MWPM) because these codes have no threshold with the default decoder (BP-OSD).

```python
rep_codes = [qldpc.codes.RepetitionCode(dist) for dist in [3, 5, 7]]
make_error_rate_figure(rep_codes, with_MWPM=True)
plt.show()
```

```python
surface_codes = [qldpc.codes.SurfaceCode(dist) for dist in [3, 5, 7]]
make_error_rate_figure(surface_codes, with_MWPM=True)
plt.show()
```

## Bivariate bicycle codes

```python
from sympy.abc import x, y

codes = [
    qldpc.codes.BBCode(
        {x: 6, y: 6},
        x**3 + y + y**2,
        y**3 + x + x**2,
    ),
    qldpc.codes.BBCode(
        {x: 15, y: 3},
        x**9 + y + y**2,
        1 + x**2 + x**7,
    ),
    qldpc.codes.BBCode(
        {x: 9, y: 6},
        x**3 + y + y**2,
        y**3 + x + x**2,
    ),
]
make_error_rate_figure(codes)
plt.show()
```
